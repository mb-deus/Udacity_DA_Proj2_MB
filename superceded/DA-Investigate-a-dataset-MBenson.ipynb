```python
# For ease of viewing, I will transpose the data frame to have countries in the columns.
df_lib_t = df_lib.T
```

```python
# Removal of countries not in the summary listing, and a quick check of the output.
df_lib_ts = df_lib_t.drop(columns=[col for col in df_lib_t if col not in countries])
df_lib_ts.head(3)
```

```python
#Plot the summary countries
df_lib_ts.plot(kind='line', figsize=(8,8))
plt.xlabel('Year');
plt.ylabel('Liberty score');
plt.title('Liberty scores for selected countries\n(2006 and 2018)');
plt.legend(fontsize=12);
```

#### Hypothesis from initial visual: 

>Between 2006 and 2008 civil liberties appear to have declined for our summary listing of countries.

#### Question: 
>Is this apparent in our total population data set?

```python
# Compare summary metrics, and change in those metrics, across first period (2006) and last period (2018)

df_lib_06 = df_lib['2006'].describe()
df_lib_18 = df_lib['2018'].describe()
stats_lib = pd.DataFrame({'metrics_06': df_lib_06, 'metrics_18': df_lib_18})
stats_lib['change'] = stats_lib['metrics_18'] - stats_lib['metrics_06']
stats_lib
```

#### Insights from the statistical analysis:  

>All statistical metrics (other than maxiumum and minimum scores) show a decrease in the liberty score across the total population.

#### Further analysis required...

>Whilst the minimum score (zero) and the maximum score (one) have not changed, we do not know if there are more or less countries achieving these scores. For example: in 2006 there may have been 10 countries acheiving a maximum score of one, and in 2018 this may have reduced to just one country.  The maximum score in both years is still one.

>The reduction in the mean would indicate the maximum and minimum scores demonstrate a negative movement but this would need to be confirmed (see below).

```python
# Which countries had a liberty score of 1 in 2006?

df_lib[df_lib['2006'] == 1]
```

```python
# Which countries had a liberty score of 1 in 2018?

df_lib[df_lib['2018'] == 1]
```

#### Question:

> Can further insight be obtained by presenting this visually?

#### Answer:

> If we group the countries into buckets based on their scores, and then count the number of countries in each bucket for both 2006 and 2018 this may show further evidence of a movement in civil liberties

```python
# Allow easy adjustment of the number of buckets in order to assess most user-friendly visual.

#Set a variable for the number of buckets which can then be changed for alternative analysis
buckets = 5

bin_edges = [-0.00001]
bin_names = []

# Populate the bin_edges and bin_names based on the number of buckets
for i in range(buckets):
    bin_edges.append((i+1)/buckets)
    bin_names.append(str((i)/buckets) + " - " + str((i+1)/buckets))
    
# Create new data columns that state which buckets the liberty score data falls into in each row    
df_lib['06bucket']= pd.cut(df_lib['2006'], bin_edges, labels=bin_names)
df_lib['18bucket']= pd.cut(df_lib['2018'], bin_edges, labels=bin_names)
```

```python
# Set variables for the data to be presented visually.  In this instance value counts of each of the buckets.

bar_1 = df_lib['06bucket'].value_counts()[bin_names]
bar_2 = df_lib['18bucket'].value_counts()[bin_names]
```

```python
# Plot the graph

plt.style.use('ggplot')
plt.figure(figsize=(8,8))
x = np.arange(buckets)
bar_width = 0.4
plt.bar(x, bar_1, width=bar_width, label='2006', color='blue' )
plt.bar(x+bar_width*0.7, bar_2, width=bar_width, label='2018', color='teal', alpha=0.8 )
plt.xlabel('Liberty score bucket (0-1)');
plt.xticks(np.arange(buckets),bin_names)
plt.ylabel('Number of countries in each bucket');
plt.title('Number of countries falling into Liberty score buckets.\n 2006 vs 2018');
plt.legend(fontsize=20);
```

#### Insight

> The most noticeable reduction for the 2006 bucket occurs in the '0.8-1.0' bucket indicating less countries achieve the highest scores.  These appear to have shifted more into the lower buckets of '0.6-0.8' and '0.2-0.4'.

### EDA: Data set 2 - GDP

```python
# Transpose the data and then extract only the countries for our summary data set

df_gdp_t = df_gdp.T
df_gdp_ts = df_gdp_t.drop(columns=[col for col in df_gdp_t if col not in countries])
df_gdp_ts.head(2)
```

### Which countries have shown the most growth over the last 100 years? (1913 - 2013)

In order to answer this we will look at both the absolute growth in GDP and the compound annual growth rate (CAGR) over the period.

```python
# Calculate the two metrics to consider

df_gdp['CAGR'] = ((df_gdp['2013']/df_gdp['1913'])**(1/100)-1) * 100
df_gdp['Abs_gth'] = df_gdp['2013']-df_gdp['1913']
```

```python
# Summarise the listing of countries for our summary selection

df_gdp_s = df_gdp.drop(index=[ind for ind in df_gdp.index if ind not in countries])
```

```python
# Sort the data based on CAGR

df_gdp_s.sort_values(by=['CAGR'], inplace=True, ascending=False)
df_gdp_s[['1913','2013','CAGR', 'Abs_gth']]
```

```python
# Sort the data based on Absolute growth

df_gdp_s.sort_values(by=['Abs_gth'], inplace=True, ascending=False)
df_gdp_s[['1913','2013','CAGR', 'Abs_gth']] 
```

```python
# Plot the data points on a scatter graph for visual assessment

x1 = df_gdp_s['CAGR']   
x2 = df_gdp_s['Abs_gth']
plt.plot(x1, x2, 'o', color='blue');
plt.xlabel('CAGR')
plt.ylabel('Abs_gth')
for x,y,z in zip(list(x1),list(x2),list(df_gdp_s.index)):
    if z in ['Germany', 'France', 'Australia', 'Russia', 'Japan']:
        plt.annotate(z, (x,y), textcoords="offset points", xytext=(3,10), ha='left') 
    elif z in ['United Kingdom']:
        plt.annotate(z, (x,y), textcoords="offset points", xytext=(3,-20), ha='left')
    else:
        plt.annotate(z, (x,y), textcoords="offset points", xytext=(3,-20), ha='right')
plt.title('GDP growth over the last century.\n CAGR vs Absolute growth');
```
