```python
# Transpose the data, and review

df_kid_t = df_kid.T
df_kid_t.head(2)
```

```python
df_kid_t.plot(kind='line', figsize=(12,12), legend=None)
```

```python
# Hash his one out in case we need it later, but ont run for now.
# df_kid_t.loc['1850':'2100',:].plot(kind='line', figsize=(12,12), legend=None)
```

```python
# Plot the summary data

countries = ['Australia', 'United Kingdom', 'United States', 'UK', 'US', 'China', 'Russia', 'France', 'Germany', 'Japan']
df_kid_ts = df_kid_t.drop(columns=[col for col in df_kid_t if col not in countries])
df_kid_ts.loc['1850':'2020',:].plot(kind='line', figsize=(12,12))
```

```python
#Rename the Uk column header so that it doesnt conflict with the US column in the next step where they are renamed
df_kid_ts = df_kid_ts.rename(columns = {'United Kingdom': 'UK', 'United States': 'US'})
```

```python
# Calculate the rolling averages and drop the first 9 rows with Nans

yrs = 10
for c in df_kid_ts.columns:
    df_kid_ts[c[0:2] + '_MA' + str(yrs)] = df_kid_ts[c].rolling(yrs).mean()
df_kid_ts.drop(df_kid_ts.index[0:yrs-1], inplace=True, axis=0)
df_kid_ts.head(15)
```

```python
df_kid_tsm = df_kid_ts.drop(columns=[col for col in df_kid_ts if col in countries])
df_kid_tsm.loc['1850':'2020',:].plot(kind='line', figsize=(12,12))
```

```python
df_kid_tsm.head(2)
```

```python
# Calculate the averages for our selection of moving averages

df_kid_tsm['sel_MA10'] = df_kid_tsm.mean(axis=1)
```

```python
# Calculate the averages for our original total population

years = 10
df_kid_t['pop_avg'] = df_kid_t.mean(axis=1)
df_kid_t['pop_MA10'] = df_kid_t['pop_avg'].rolling(yrs).mean()
```

```python
# Add on the new column to our existing dataset

df_kid_tsm = df_kid_tsm.assign(pop_MA10=df_kid_t['pop_MA10'])
```

```python
# And check...

df_kid_tsm.head(2)
```

```python
# And plot

from1 = str(1850)
to1 = str(2020)

df_kid_tsm.loc[from1:to1,:].plot(kind='line', figsize=(12,12))

xx1 = df_kid_tsm.loc[from1:to1,:].index
yy1 = list(df_kid_tsm['sel_MA10'].loc[from1:to1])
zz1 = list(df_kid_tsm['pop_MA10'].loc[from1:to1])

plt.title("Number of children per woman")
plt.ylabel("Number_of_children")
plt.xlabel("Year")
plt.plot(xx1,yy1,label = 'AVG_MA10', lw=5, ls='--')
plt.plot(xx1,zz1,label = 'AVG_MA10', lw=5, ls='--')
plt.show()
```
